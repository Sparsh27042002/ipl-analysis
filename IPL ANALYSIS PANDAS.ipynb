```python
vk=delivery[delivery['batsman']=='V Kohli']
```

```python
def run_scored(batsman_name):
   vk=delivery[delivery['batsman']==batsman_name]
   return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
run_scored('Yuvraj Singh')
```

# find the most destructive death over batsman in the history of IPL
Strike rate=(Number of runs/number of balls)/100
Min 200 balls to be played between 16-20 overs

```python
death_overs=delivery[delivery['over']>15]
```

```python
all_batsman=death_overs.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
batsman_list
```

```python
#runs scored by all these 43 batsman
#balls played by these 43 batsman
final=delivery[delivery['batsman'].isin(batsman_list)]
final
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
runs
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
balls
```

```python
sr=(runs/balls)*100
```

```python
sr.sort_values(ascending=False).head()
```
